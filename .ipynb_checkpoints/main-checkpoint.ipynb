# Types of routing
## Map Initialisation

```python
import folium

# Center coordinates for the map
center_coordinates = [53.36486137451511, -1.8160056925378616]  # Example: Central London

tiles = "https://api.os.uk/maps/raster/v1/zxy/Leisure_27700/{z}/{x}/{y}.png?key=GGF16a7QH0WAQEkXEXqCwHwxLp0aoMAL"

tile_backup = "opentopomap"

# Initialize the map using OS Maps as the default tiles
m = folium.Map(location=center_coordinates, zoom_start=13, tiles=tile_backup, max_zoom=20, min_zoom=0, attr="© Ordnance Survey")


folium.WmsTileLayer(
    url="https://api.os.uk/maps/raster/v1/wmts?key=GGF16a7QH0WAQEkXEXqCwHwxLp0aoMAL",
    name="Leisure_27700",
    fmt="image/png",
    styles="default",
    layers="Leisure_27700",
    attr=u"© Ordnance Survey",
    request='GetTile',
    tileCol='{x}',
    tileRow='{y}',
    tileMatrix='{z}',
    transparent=True,
    overlay=True,
    control=True,
).add_to(m)

folium.LayerControl().add_to(m)
```

```python
import GA
coords = GA.funny()
```

```python
folium.PolyLine(
    locations=coords,
    color="#FF0000",
    weight=5,
    tooltip="GA",
).add_to(m)
```
